## Intro to coding

This Jupyter notebook will walk you through the basics of getting started with coding and help you use code to make doing your project easier.

First, let's play around with the notebook environment a bit. To run the following chunk of code, click inside the code cell and press Shift+Enter.

```python
# Make some variables called x and y and assign them some values
x = 1
y = 2
# Do some basic math with those variables
z = x + y
# See what the output is
print(z)
```

The above snippet of code introduces a few key ideas:
- variable assignment
- print statements

When we are coding, we can use variables to store information with any name that we like. In this case, I've chosen x and y. We can then *assign* a piece of information to those variables using an "=" sign. Here, we've chosen to use them to store some numbers. We can do different manipulations---like mathematical operations---with the variables. Print statements are a very helpful way to see the output of those manipulations.

I have also used comments (the green sections of text) to explain what the code is doing. Anything in the code that has a "#" in front of it will be "commented out" and will not actually process as code.

In this simple example, it may not seem very useful to use variables. We could have just typed `print(1+2)` and gotten the same result. However, the advantage of using variables is that it makes it much easier to edit our code later on if we decide we want to change something.

For example:

```python
# Make some new variables called a and b
a = 1725
b = 9823
# Do a couple of mathematical operations with the variables
c = a + b
d = a - b
f = a * b
g = a / b
# See what the output is
print("c equals", c)
print("d equals", d)
print("f equals", f)
print("g equals", g)
```

Try changing the value of a and/or b in the above example and running it again. You can see the results of all four operations---addition, subtraction, multiplication, and division---while only having to type in your new value once!

## Using packages and working with arrays

In the coding world, a lot of people will write code to do something useful and then make it available to you to use. We often start our Python code by importing a bunch of these packages which will be useful for us.

```python
# NumPy (Numerical Python) is a package that allows us to do a bunch of mathematical operations
# We will import it and call it "np" for short
import numpy as np
# Matplotlib is a package that allows us to make graphs to easily visualize the information we are working with
# We will import it and call it "plt" for short
import matplotlib.pyplot as plt
```

Let's try using some of the "functions" that come built in with NumPy.

```python
# We will use our variables x and y from above
# We do not have to define them again since we are working in the same notebook still!
sine_value = np.sin(x)
cosine_value = np.cos(y)
# Look at the output
print("Sine value of x is", sine_value)
print("Cosine value of y is", cosine_value)
```

We can also use NumPy to work with arrays, which are basically lists of data. This can make our coding more efficient.

For example:

```python
# Define an array
array = np.array([1, 2, 3, 4, 5])
# Add 3 to each number in the array
array_plus_three = array + 3
# Look at the output
print(array_plus_three)
```

We were able to add 3 to each number in the array with just one line of code!

Arrays are also useful for making graphs because we can use them to store a bunch of data points.

```python
# Make an input array called x (this will overwrite the earlier code where I made x = 1)
x = np.array([1, 2, 3, 4, 5])
# Use NumPy to perform a function on the array
y = 2 * x + 10
# Use the plotting package Matplotlib to graph the data
plt.plot(x, y)
```

## Reading in data from another file

For your project, I will have you store information about your galaxies in a spreadsheet. We will then be able to use NumPy to import that data into our code. Let's see how this works on an example:

```python
# Give Colab access to Google Drive files
from google.colab import drive
drive.mount("/content/drive")
# Tell Colab to look in the folder called "Galaxy Exploration (2024)" which lives inside a folder called "WSP"
%cd /content/drive/My Drive/WSP/Galaxy Exploration (2024)
# Load in the data from our example file
data = np.loadtxt("Example.csv", dtype="int,int", delimiter=",", usecols=(0, 1), unpack=True)
# Look at the output
print(data)
```

You can see that we have loaded in two arrays: one corresponds to the first column from the file, and one corresponds to the second column from the file. Let's plot them.

```python
# Assign variables x and y to represent the arrays we loaded in
# Note that in Python, everything is zero-indexed, which means that the first array will be called "data[0]"
x = data[0]
y = data[1]
# Graph the data
plt.plot(x, y)
# Add some labels to the graph so it is easier to read
plt.xlabel("x")
plt.ylabel("y")
```
